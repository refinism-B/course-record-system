```python
from mod import O_general as gr
from mod.O_config import GROUP_LIST, MANAGER_DICT
from mod.O_general import Student
```

```python
# step 0: 建立連線，並將所有分頁讀入
client = gr.create_client()
df_dict = gr.connect_and_read_all_sheets(client=client)
```

```python
# step 1: 負責人登入，並選擇查看的小組
user = input("請輸入使用者名稱") # 提供選項["林玲", "桂丹", "善美", 總覽]
group = input("請輸入查詢小組") # 提供選項["社青小組", "善牧小組"]
if user in list(MANAGER_DICT) and group in GROUP_LIST:
    course = MANAGER_DICT[user]
    df = df_dict[group]

    mask = (df["課程"] == course)
    df_temp = df[mask]
else:
    print("使用者名稱或小組名稱輸入錯誤")
```

```python
# step 1-2: 將符合條件的成員列出
df_temp
```

```python
# step 2: 輸入修改對象
search_name = input("請輸入欲修改的對象") # 使用st.multiselect方式搜尋所有姓名

mask = (df_temp["姓名"] == search_name)
result = df_temp[mask]
result_idx = result.index[0]

search_student = Student.from_series(df.loc[result_idx])
```

```python
df.loc[result_idx]
```

```python
# step 3: 選擇修改月份
update_month = input("請輸入欲修改的月份") # 提供選項：{"執行狀況：一月": jan, "執行狀況：二月": feb, "執行狀況：三月": mar, "執行狀況：四月": apr}

comment = input("請輸入說明") # 可輸入任意字串，上限100字

setattr(search_student, update_month, comment)
```

```python
# step 4: 將修改後資料存回
df.loc[result_idx] = search_student.to_dict()
```

```python
worksheet = gr.connect_sheet_file(client=client, sheet_name=group)

gr.save_to_sheet(worksheet=worksheet, df=df)
```
